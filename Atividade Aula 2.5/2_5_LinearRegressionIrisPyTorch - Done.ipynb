# Regressão Linear com PyTorch

## Objetivo

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se
- o cálculo do gradiente automático do PyTorch, porém
- a função de perda e
- atualização dos parâmetros é feita de forma explícita.
- A rede é criada com uma camada nn.Linear()

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinear_Otimizacao.png)

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
# Inicialização usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform_(model.weight.data, -0.1, 0.1)
model.weight
```

### Testando o predict da rede

```python
xx = torch.ones((5,2))
xx
```

```python
model(Variable(xx))
```

## Treinamento

```python
num_epochs = 100
learning_rate = 0.5

w0_list = []
w0_gradient_list = []
w1_list = []
w1_gradient_list = []

for epoch in range(num_epochs):
    #inputs = Variable(x_train_bias)
    #target = Variable(y_train)
    inputs = x_train_bias
    target = y_train

    w0_list.append(model.weight.data[0,0].item())
    w1_list.append(model.weight.data[0,1].item())

    # forward - predict
    y_pred = model(inputs)

    # loss cálculo da função de perda
    loss = torch.mean(torch.square(y_pred - target))

    # backward propagation - cálculo gradient
    loss.backward()

    w0_gradient_list.append(model.weight.grad.data[0,0].item())
    w1_gradient_list.append(model.weight.grad.data[0,1].item())

    # gradiente descendente
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # zeragem parâmetros próximo cálculo do gradiente
    model.weight.grad.data.zero_()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

### Cálculo da perda ao fazer predict com os presos já treinados

```python
with torch.no_grad():
  y_pred = model(x_train_bias)
  loss = torch.mean(torch.square(y_pred - y_train))
  print('loss: {:.6f}'.format(loss.data))
```

### Evolução dos parâmetros w0 e w1

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

epochs = range(num_epochs)

ax1.plot(epochs, w0_list, label='Valor do Peso (w0)', color='blue')
ax1.set_ylabel('Valor do Peso (w0)')
ax1.set_title('Convergência do Peso (w0) e seu Gradiente ao Longo das Épocas')
ax1.legend()
ax1.grid(True)

ax2.plot(epochs, w0_gradient_list, label='Gradiente de w0 (grad_w0)', color='red')
ax2.set_xlabel('Época')
ax2.set_ylabel('Valor do Gradiente')

ax2.axhline(0, color='black', linestyle='--', linewidth=0.8, label='Gradiente = 0')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

epochs = range(num_epochs)

ax1.plot(epochs, w1_list, label='Valor do Peso (w1)', color='blue')
ax1.set_ylabel('Valor do Peso (w1)')
ax1.set_title('Convergência do Peso (w1) e seu Gradiente ao Longo das Épocas')
ax1.legend()
ax1.grid(True)

ax2.plot(epochs, w1_gradient_list, label='Gradiente de w1 (grad_w1)', color='red')
ax2.set_xlabel('Época')
ax2.set_ylabel('Valor do Gradiente')

ax2.axhline(0, color='black', linestyle='--', linewidth=0.8, label='Gradiente = 0')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
